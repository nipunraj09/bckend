```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/kaggle/input/plastic-pollution/1- global-plastics-production.csv')
data.head()
```

```python
x=data['Annual plastic production between 1950 and 2019']
y=data['Year']
plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
from scipy import stats



slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
data = data['Annual plastic production between 1950 and 2019'].to_frame()
 
# calculating exponential moving average
# using .ewm(span).mean() , with window size = 30
data['EWMA'] = data['Annual plastic production between 1950 and 2019'].ewm(span=30).mean()
data
```

```python
data[['Annual plastic production between 1950 and 2019', 'EWMA']].plot(label='Net pollution',
                                   figsize=(16, 8))
```
